```python
# ============================================================
# G4: CUSTOM CNN WITH DROPBLOCK
# Rubber Leaf Disease Classification
# Study: Dropout Implementation in CNN for Rubber Leaf Disease Classification
# Model Type: Custom CNN with DropBlock2D
# Platform: Google Colab
# ============================================================


# ============================================================
# 1. IMPORTS
# ============================================================

import os
import time
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import (
    EarlyStopping,
    ModelCheckpoint,
    ReduceLROnPlateau,
    CSVLogger
)

from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

import cv2
import matplotlib.cm as cm

print("TensorFlow version:", tf.__version__)


# ============================================================
# 2. REPRODUCIBILITY SETTINGS
# ============================================================

SEED = 42

random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)
os.environ["PYTHONHASHSEED"] = str(SEED)


# ============================================================
# 3. GOOGLE DRIVE MOUNTING
# ============================================================

from google.colab import drive
drive.mount("/content/drive")


# ============================================================
# 4. DATASET AND EXPERIMENT CONFIGURATION
# ============================================================

DATASET_DIR = "/content/drive/MyDrive/DIT/Dissertation/rubber"

TRAIN_DIR = os.path.join(DATASET_DIR, "training")
VAL_DIR = os.path.join(DATASET_DIR, "validation")
TEST_DIR = os.path.join(DATASET_DIR, "testing")

IMG_HEIGHT = 224
IMG_WIDTH = 224
IMG_SIZE = (IMG_HEIGHT, IMG_WIDTH)
CHANNELS = 3

BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 1e-4

# ------------------------------------------------------------
# G4 Experimental Variables
# Recommended dropout rates: 0.20, 0.30, 0.40, 0.50
# Recommended block size: 3, 5, or 7
# ------------------------------------------------------------

DROPBLOCK_RATE = 0.20
DROPBLOCK_BLOCK_SIZE = 3

EXPERIMENT_GROUP = "G4"
MODEL_NAME = f"G4_Custom_CNN_DropBlock_rate_{DROPBLOCK_RATE}_block_{DROPBLOCK_BLOCK_SIZE}"

OUTPUT_DIR = (
    f"/content/drive/MyDrive/DIT/Dissertation/rubber/"
    f"g4_dropblock_rate_{DROPBLOCK_RATE}_block_{DROPBLOCK_BLOCK_SIZE}"
)

os.makedirs(OUTPUT_DIR, exist_ok=True)

MODEL_SAVE_PATH = os.path.join(
    OUTPUT_DIR,
    f"{MODEL_NAME}.keras"
)

BEST_MODEL_SAVE_PATH = os.path.join(
    OUTPUT_DIR,
    f"best_{MODEL_NAME}.keras"
)

TRAINING_LOG_PATH = os.path.join(
    OUTPUT_DIR,
    "training_log.csv"
)

RESULTS_JSON_PATH = os.path.join(
    OUTPUT_DIR,
    "evaluation_results.json"
)

CLASSIFICATION_REPORT_PATH = os.path.join(
    OUTPUT_DIR,
    "classification_report.csv"
)

CONFUSION_MATRIX_PATH = os.path.join(
    OUTPUT_DIR,
    "confusion_matrix.csv"
)

SUMMARY_TABLE_PATH = os.path.join(
    OUTPUT_DIR,
    "g4_dropblock_summary_table.csv"
)

GRADCAM_OUTPUT_DIR = os.path.join(
    OUTPUT_DIR,
    "gradcam_outputs"
)

os.makedirs(GRADCAM_OUTPUT_DIR, exist_ok=True)


# ============================================================
# 5. DATASET STRUCTURE CHECKING
# ============================================================

def check_dataset_directory(path, folder_name):
    """
    Checks whether a required dataset folder exists.
    """

    if not os.path.exists(path):
        raise FileNotFoundError(
            f"{folder_name} directory not found: {path}\n"
            "Please check your dataset path and folder structure."
        )

    print(f"{folder_name} directory found: {path}")


check_dataset_directory(TRAIN_DIR, "training")
check_dataset_directory(VAL_DIR, "validation")
check_dataset_directory(TEST_DIR, "testing")


# ============================================================
# 6. DATASET LOADING
# ============================================================

train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_DIR,
    labels="inferred",
    label_mode="int",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    VAL_DIR,
    labels="inferred",
    label_mode="int",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    TEST_DIR,
    labels="inferred",
    label_mode="int",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds.class_names
NUM_CLASSES = len(class_names)

print("\nDetected Classes:")
for index, class_name in enumerate(class_names):
    print(f"{index}: {class_name}")

print("\nNumber of classes:", NUM_CLASSES)


# ============================================================
# 7. DATA PIPELINE OPTIMIZATION
# ============================================================

AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000, seed=SEED).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)


# ============================================================
# 8. DATA AUGMENTATION LAYER
# ============================================================

data_augmentation = tf.keras.Sequential(
    [
        layers.RandomFlip("horizontal", name="random_horizontal_flip"),
        layers.RandomRotation(0.10, name="random_rotation"),
        layers.RandomZoom(0.10, name="random_zoom"),
        layers.RandomContrast(0.10, name="random_contrast"),
    ],
    name="data_augmentation"
)


# ============================================================
# 9. CUSTOM DROPBLOCK2D LAYER - FIXED FOR KERAS 3 / COLAB
# ============================================================

@tf.keras.utils.register_keras_serializable()
class DropBlock2D(layers.Layer):
    """
    DropBlock2D regularization layer.

    DropBlock removes contiguous square regions from feature maps
    during training. During inference, the layer returns the input unchanged.

    Parameters:
    drop_rate : float
        Probability of dropping activations.
    block_size : int
        Size of the square block to remove.
    """

    def __init__(self, drop_rate=0.30, block_size=5, **kwargs):
        super(DropBlock2D, self).__init__(**kwargs)
        self.drop_rate = float(drop_rate)
        self.block_size = int(block_size)

    def call(self, inputs, training=False):
        """
        Applies DropBlock only during training.
        During inference or model construction, the input is returned unchanged.
        """

        if (not training) or self.drop_rate == 0.0:
            return inputs

        input_shape = tf.shape(inputs)

        batch_size = input_shape[0]
        height = input_shape[1]
        width = input_shape[2]
        channels = input_shape[3]

        block_size = tf.minimum(
            self.block_size,
            tf.minimum(height, width)
        )

        block_size_float = tf.cast(block_size, inputs.dtype)
        height_float = tf.cast(height, inputs.dtype)
        width_float = tf.cast(width, inputs.dtype)

        gamma = (
            self.drop_rate
            * height_float
            * width_float
            / tf.square(block_size_float)
            / (
                (height_float - block_size_float + 1.0)
                * (width_float - block_size_float + 1.0)
            )
        )

        random_tensor = tf.random.uniform(
            shape=(batch_size, height, width, channels),
            dtype=inputs.dtype
        )

        sampling_mask = tf.cast(
            random_tensor < gamma,
            inputs.dtype
        )

        block_mask = tf.nn.max_pool2d(
            sampling_mask,
            ksize=self.block_size,
            strides=1,
            padding="SAME"
        )

        block_mask = 1.0 - block_mask

        normalizer = (
            tf.cast(tf.size(block_mask), inputs.dtype)
            / (tf.reduce_sum(block_mask) + tf.keras.backend.epsilon())
        )

        outputs = inputs * block_mask * normalizer

        return outputs

    def compute_output_shape(self, input_shape):
        """
        Output shape is the same as input shape.
        """

        return input_shape

    def get_config(self):
        """
        Enables model saving and loading with the custom layer.
        """

        config = super(DropBlock2D, self).get_config()
        config.update({
            "drop_rate": self.drop_rate,
            "block_size": self.block_size
        })
        return config


# ============================================================
# 10. G4 CUSTOM CNN MODEL WITH DROPBLOCK
# ============================================================

def build_g4_dropblock_cnn(
    input_shape=(224, 224, 3),
    num_classes=3,
    dropblock_rate=0.20,
    dropblock_block_size=3
):
    """
    Builds G4 Custom CNN with DropBlock2D.

    This model follows the same base architecture as G1, G2, and G3.
    The experimental intervention is DropBlock2D applied after deeper
    convolutional blocks.

    G4 placement:
        DropBlock after Block 3
        DropBlock after Block 4

    Rationale:
        Earlier layers capture low-level features such as edges and veins.
        Deeper layers capture more disease-relevant features such as
        spots, lesions, necrotic patches, and discoloration regions.
    """

    inputs = layers.Input(shape=input_shape, name="input_image")

    # Data augmentation is active during training only.
    x = data_augmentation(inputs)

    # Pixel normalization.
    x = layers.Rescaling(1.0 / 255, name="rescaling")(x)

    # --------------------------------------------------------
    # Convolutional Block 1
    # Output after pooling: 112 x 112 x 32
    # --------------------------------------------------------
    x = layers.Conv2D(
        32,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv1_1"
    )(x)
    x = layers.BatchNormalization(name="bn1_1")(x)

    x = layers.Conv2D(
        32,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv1_2"
    )(x)
    x = layers.BatchNormalization(name="bn1_2")(x)

    x = layers.MaxPooling2D(pool_size=(2, 2), name="pool1")(x)

    # --------------------------------------------------------
    # Convolutional Block 2
    # Output after pooling: 56 x 56 x 64
    # --------------------------------------------------------
    x = layers.Conv2D(
        64,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv2_1"
    )(x)
    x = layers.BatchNormalization(name="bn2_1")(x)

    x = layers.Conv2D(
        64,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv2_2"
    )(x)
    x = layers.BatchNormalization(name="bn2_2")(x)

    x = layers.MaxPooling2D(pool_size=(2, 2), name="pool2")(x)

    # --------------------------------------------------------
    # Convolutional Block 3
    # Output after pooling: 28 x 28 x 128
    # DropBlock begins here.
    # --------------------------------------------------------
    x = layers.Conv2D(
        128,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv3_1"
    )(x)
    x = layers.BatchNormalization(name="bn3_1")(x)

    x = layers.Conv2D(
        128,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv3_2"
    )(x)
    x = layers.BatchNormalization(name="bn3_2")(x)

    x = layers.MaxPooling2D(pool_size=(2, 2), name="pool3")(x)

    x = DropBlock2D(
        drop_rate=dropblock_rate,
        block_size=dropblock_block_size,
        name=f"dropblock_block3_rate_{dropblock_rate}_size_{dropblock_block_size}"
    )(x)

    # --------------------------------------------------------
    # Convolutional Block 4
    # Output after pooling: 14 x 14 x 256
    # --------------------------------------------------------
    x = layers.Conv2D(
        256,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv4_1"
    )(x)
    x = layers.BatchNormalization(name="bn4_1")(x)

    x = layers.Conv2D(
        256,
        kernel_size=(3, 3),
        padding="same",
        activation="relu",
        name="conv4_2"
    )(x)
    x = layers.BatchNormalization(name="bn4_2")(x)

    x = layers.MaxPooling2D(pool_size=(2, 2), name="pool4")(x)

    x = DropBlock2D(
        drop_rate=dropblock_rate,
        block_size=dropblock_block_size,
        name=f"dropblock_block4_rate_{dropblock_rate}_size_{dropblock_block_size}"
    )(x)

    # --------------------------------------------------------
    # Classification Head
    # --------------------------------------------------------
    x = layers.GlobalAveragePooling2D(name="global_average_pooling")(x)

    x = layers.Dense(
        128,
        activation="relu",
        name="dense_128"
    )(x)

    outputs = layers.Dense(
        num_classes,
        activation="softmax",
        name="softmax_output"
    )(x)

    model = models.Model(
        inputs=inputs,
        outputs=outputs,
        name=MODEL_NAME
    )

    return model


model = build_g4_dropblock_cnn(
    input_shape=(IMG_HEIGHT, IMG_WIDTH, CHANNELS),
    num_classes=NUM_CLASSES,
    dropblock_rate=DROPBLOCK_RATE,
    dropblock_block_size=DROPBLOCK_BLOCK_SIZE
)

model.summary()


# ============================================================
# 11. MODEL COMPILATION
# ============================================================

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)


# ============================================================
# 12. CALLBACKS
# ============================================================

callbacks = [
    EarlyStopping(
        monitor="val_loss",
        patience=30,
        restore_best_weights=True,
        verbose=1
    ),

    ModelCheckpoint(
        filepath=BEST_MODEL_SAVE_PATH,
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    ),

    ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.2,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),

    CSVLogger(
        filename=TRAINING_LOG_PATH,
        append=False
    )
]

# ============================================================
# COUNT IMAGES PER CLASS IN TRAIN, VAL, AND TEST FOLDERS
# ============================================================

from pathlib import Path

def count_images_per_class(directory):
    directory = Path(directory)
    valid_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".gif"]

    print(f"\n================ {directory.name.upper()} SET ================")

    for class_folder in sorted(directory.iterdir()):
        if class_folder.is_dir():
            count = sum(
                1 for file in class_folder.rglob("*")
                if file.suffix.lower() in valid_extensions
            )
            print(f"{class_folder.name}: {count}")

count_images_per_class(TRAIN_DIR)
count_images_per_class(VAL_DIR)
count_images_per_class(TEST_DIR)

# ============================================================
# COUNT IMAGES PER CLASS IN TRAIN, VAL, AND TEST FOLDERS
# ============================================================

from pathlib import Path

def count_images_per_class(directory):
    directory = Path(directory)
    valid_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".gif"]

    print(f"\n================ {directory.name.upper()} SET ================")

    for class_folder in sorted(directory.iterdir()):
        if class_folder.is_dir():
            count = sum(
                1 for file in class_folder.rglob("*")
                if file.suffix.lower() in valid_extensions
            )
            print(f"{class_folder.name}: {count}")

count_images_per_class(TRAIN_DIR)
count_images_per_class(VAL_DIR)
count_images_per_class(TEST_DIR)


# ============================================================
# 13. MODEL TRAINING
# ============================================================

# ============================================================
# CLASS WEIGHT COMPUTATION
# ============================================================

from sklearn.utils.class_weight import compute_class_weight

train_labels = []

for _, labels in train_ds.unbatch():
    train_labels.append(int(labels.numpy()))

train_labels = np.array(train_labels)

class_weights_array = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(train_labels),
    y=train_labels
)

class_weights = {
    i: float(class_weights_array[i])
    for i in range(len(class_weights_array))
}

print("\n================ CLASS WEIGHTS ================")
for class_index, weight in class_weights.items():
    print(f"{class_names[class_index]}: {weight:.4f}")


# start training the model

start_training_time = time.time()

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks,
    class_weight=class_weights,
    verbose=1
)

end_training_time = time.time()
training_time_seconds = end_training_time - start_training_time

print("\n================ TRAINING TIME ================")
print(f"Total Training Time: {training_time_seconds:.2f} seconds")


# ============================================================
# 14. SAVE FINAL MODEL
# ============================================================

model.save(MODEL_SAVE_PATH)

print("\n================ MODEL SAVED ================")
print(f"Final model saved to: {MODEL_SAVE_PATH}")
print(f"Best model saved to : {BEST_MODEL_SAVE_PATH}")


# ============================================================
# 15. TRAINING HISTORY VISUALIZATION
# ============================================================

def plot_training_history(history, output_dir):
    """
    Plots and saves training and validation accuracy and loss curves.
    """

    history_dict = history.history

    plt.figure(figsize=(8, 6))
    plt.plot(history_dict["accuracy"], label="Training Accuracy")
    plt.plot(history_dict["val_accuracy"], label="Validation Accuracy")
    plt.title(f"{MODEL_NAME} Training and Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)

    accuracy_plot_path = os.path.join(output_dir, "accuracy_curve.png")
    plt.savefig(accuracy_plot_path, dpi=300, bbox_inches="tight")
    plt.show()

    plt.figure(figsize=(8, 6))
    plt.plot(history_dict["loss"], label="Training Loss")
    plt.plot(history_dict["val_loss"], label="Validation Loss")
    plt.title(f"{MODEL_NAME} Training and Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)

    loss_plot_path = os.path.join(output_dir, "loss_curve.png")
    plt.savefig(loss_plot_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Accuracy curve saved to: {accuracy_plot_path}")
    print(f"Loss curve saved to    : {loss_plot_path}")


plot_training_history(history, OUTPUT_DIR)


# ============================================================
# 16. HELPER FUNCTION: TRUE LABELS AND PREDICTIONS
# ============================================================

def get_predictions(model, dataset):
    """
    Generates true labels, predicted labels, and prediction probabilities.
    """

    y_true = []
    y_pred = []
    y_prob = []

    for images, labels in dataset:
        probabilities = model.predict(images, verbose=0)
        predictions = np.argmax(probabilities, axis=1)

        y_true.extend(labels.numpy())
        y_pred.extend(predictions)
        y_prob.extend(probabilities)

    return np.array(y_true), np.array(y_pred), np.array(y_prob)


# ============================================================
# 17. TEST SET EVALUATION
# ============================================================

y_true, y_pred, y_prob = get_predictions(model, test_ds)

# ============================================================
# 16.1 DIAGNOSTIC: TRUE VS PREDICTED CLASS DISTRIBUTION
# ============================================================

unique_true, true_counts = np.unique(y_true, return_counts=True)
unique_pred, pred_counts = np.unique(y_pred, return_counts=True)

print("\n================ TRUE CLASS DISTRIBUTION ================")
for class_index, count in zip(unique_true, true_counts):
    print(f"{class_names[class_index]}: {count}")

print("\n================ PREDICTED CLASS DISTRIBUTION ================")
for class_index, count in zip(unique_pred, pred_counts):
    print(f"{class_names[class_index]}: {count}")


# ============================================================
# 16.2 DIAGNOSTIC: SAMPLE PREDICTION PROBABILITIES
# ============================================================

print("\n================ SAMPLE PREDICTION PROBABILITIES ================")

num_samples_to_inspect = min(10, len(y_true))

for i in range(num_samples_to_inspect):
    print(f"\nSample {i + 1}")
    print(f"True Class      : {class_names[y_true[i]]}")
    print(f"Predicted Class : {class_names[y_pred[i]]}")
    print("Probabilities:")

    for class_index, prob in enumerate(y_prob[i]):
        print(f"  {class_names[class_index]}: {prob:.4f}")


# ============================================================
# 16.3 COMPUTE TEST METRICS
# ============================================================

test_accuracy = accuracy_score(y_true, y_pred)

test_precision_macro = precision_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

test_recall_macro = recall_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

test_f1_macro = f1_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

test_precision_weighted = precision_score(
    y_true,
    y_pred,
    average="weighted",
    zero_division=0
)

test_recall_weighted = recall_score(
    y_true,
    y_pred,
    average="weighted",
    zero_division=0
)

test_f1_weighted = f1_score(
    y_true,
    y_pred,
    average="weighted",
    zero_division=0
)

test_accuracy = accuracy_score(y_true, y_pred)

test_precision_macro = precision_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

test_recall_macro = recall_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

test_f1_macro = f1_score(
    y_true,
    y_pred,
    average="macro",
    zero_division=0
)

test_precision_weighted = precision_score(
    y_true,
    y_pred,
    average="weighted",
    zero_division=0
)

test_recall_weighted = recall_score(
    y_true,
    y_pred,
    average="weighted",
    zero_division=0
)

test_f1_weighted = f1_score(
    y_true,
    y_pred,
    average="weighted",
    zero_division=0
)

print("\n================ TEST SET PERFORMANCE ================")
print(f"Accuracy           : {test_accuracy:.4f}")
print(f"Precision Macro    : {test_precision_macro:.4f}")
print(f"Recall Macro       : {test_recall_macro:.4f}")
print(f"F1-score Macro     : {test_f1_macro:.4f}")
print(f"Precision Weighted : {test_precision_weighted:.4f}")
print(f"Recall Weighted    : {test_recall_weighted:.4f}")
print(f"F1-score Weighted  : {test_f1_weighted:.4f}")


# ============================================================
# 18. CLASSIFICATION REPORT
# ============================================================

report_dict = classification_report(
    y_true,
    y_pred,
    target_names=class_names,
    output_dict=True,
    zero_division=0
)

report_df = pd.DataFrame(report_dict).transpose()
report_df.to_csv(CLASSIFICATION_REPORT_PATH, index=True)

print("\n================ CLASSIFICATION REPORT ================")
print(report_df)
print(f"\nClassification report saved to: {CLASSIFICATION_REPORT_PATH}")


# ============================================================
# 19. CONFUSION MATRIX
# ============================================================

cmatrix = confusion_matrix(y_true, y_pred)

cmatrix_df = pd.DataFrame(
    cmatrix,
    index=class_names,
    columns=class_names
)

cmatrix_df.to_csv(CONFUSION_MATRIX_PATH, index=True)

print("\n================ CONFUSION MATRIX ================")
print(cmatrix_df)
print(f"\nConfusion matrix saved to: {CONFUSION_MATRIX_PATH}")


# ============================================================
# 20. CONFUSION MATRIX VISUALIZATION
# ============================================================

def plot_confusion_matrix(cmatrix, class_names, output_dir):
    """
    Plots and saves the confusion matrix as an image.
    """

    plt.figure(figsize=(10, 8))
    plt.imshow(cmatrix, interpolation="nearest")
    plt.title(f"{MODEL_NAME} Confusion Matrix")
    plt.colorbar()

    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45, ha="right")
    plt.yticks(tick_marks, class_names)

    threshold = cmatrix.max() / 2

    for i in range(cmatrix.shape[0]):
        for j in range(cmatrix.shape[1]):
            plt.text(
                j,
                i,
                format(cmatrix[i, j], "d"),
                ha="center",
                va="center",
                color="white" if cmatrix[i, j] > threshold else "black"
            )

    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.tight_layout()

    cmatrix_plot_path = os.path.join(output_dir, "confusion_matrix.png")
    plt.savefig(cmatrix_plot_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Confusion matrix plot saved to: {cmatrix_plot_path}")


plot_confusion_matrix(cmatrix, class_names, OUTPUT_DIR)


# ============================================================
# 21. GENERALIZATION GAP ANALYSIS
# ============================================================

best_train_accuracy = max(history.history["accuracy"])
best_val_accuracy = max(history.history["val_accuracy"])

final_train_accuracy = history.history["accuracy"][-1]
final_val_accuracy = history.history["val_accuracy"][-1]

best_generalization_gap = best_train_accuracy - best_val_accuracy
final_generalization_gap = final_train_accuracy - final_val_accuracy

print("\n================ GENERALIZATION GAP ANALYSIS ================")
print(f"Best Training Accuracy    : {best_train_accuracy:.4f}")
print(f"Best Validation Accuracy  : {best_val_accuracy:.4f}")
print(f"Best Generalization Gap   : {best_generalization_gap:.4f}")
print(f"Final Training Accuracy   : {final_train_accuracy:.4f}")
print(f"Final Validation Accuracy : {final_val_accuracy:.4f}")
print(f"Final Generalization Gap  : {final_generalization_gap:.4f}")


# ============================================================
# 22. MODEL SIZE AND COMPLEXITY ANALYSIS
# ============================================================

def get_model_size_mb(model_path):
    """
    Computes model file size in megabytes.
    """

    size_bytes = os.path.getsize(model_path)
    size_mb = size_bytes / (1024 * 1024)

    return size_mb


model_size_mb = get_model_size_mb(MODEL_SAVE_PATH)
total_params = model.count_params()

print("\n================ MODEL SIZE AND COMPLEXITY ================")
print(f"Total Parameters : {total_params:,}")
print(f"Model Size       : {model_size_mb:.2f} MB")


# ============================================================
# 23. INFERENCE SPEED ANALYSIS
# ============================================================

def measure_inference_speed(model, dataset, num_batches=10):
    """
    Measures average inference time per image and throughput.
    """

    total_images = 0
    total_time = 0.0

    # Warm-up prediction.
    for images, _ in dataset.take(1):
        _ = model.predict(images, verbose=0)

    for images, _ in dataset.take(num_batches):
        batch_size = images.shape[0]

        start_time = time.time()
        _ = model.predict(images, verbose=0)
        end_time = time.time()

        total_time += end_time - start_time
        total_images += batch_size

    avg_time_per_image = total_time / total_images
    images_per_second = total_images / total_time

    return avg_time_per_image, images_per_second


avg_inference_time, images_per_second = measure_inference_speed(
    model=model,
    dataset=test_ds,
    num_batches=10
)

print("\n================ INFERENCE SPEED ANALYSIS ================")
print(f"Average Inference Time per Image : {avg_inference_time:.6f} seconds")
print(f"Images per Second                : {images_per_second:.2f}")


# ============================================================
# 24. ROBUSTNESS TESTING
# ============================================================

def add_gaussian_noise(images, mean=0.0, stddev=0.05):
    """
    Adds Gaussian noise to image tensors.
    Images are assumed to be in pixel range 0-255.
    """

    images = tf.cast(images, tf.float32)

    noise = tf.random.normal(
        shape=tf.shape(images),
        mean=mean,
        stddev=stddev * 255.0
    )

    noisy_images = images + noise
    noisy_images = tf.clip_by_value(noisy_images, 0.0, 255.0)

    return noisy_images


def adjust_brightness(images, delta=0.15):
    """
    Applies brightness perturbation to image tensors.
    Images are assumed to be in pixel range 0-255.
    """

    images = tf.cast(images, tf.float32)

    adjusted_images = tf.image.adjust_brightness(
        images,
        delta=delta * 255.0
    )

    adjusted_images = tf.clip_by_value(adjusted_images, 0.0, 255.0)

    return adjusted_images


def evaluate_on_perturbed_dataset(model, dataset, perturbation_type="noise"):
    """
    Evaluates model robustness under image perturbations.
    """

    y_true_perturbed = []
    y_pred_perturbed = []

    for images, labels in dataset:

        if perturbation_type == "noise":
            perturbed_images = add_gaussian_noise(images)

        elif perturbation_type == "brightness":
            perturbed_images = adjust_brightness(images)

        else:
            perturbed_images = images

        probabilities = model.predict(perturbed_images, verbose=0)
        predictions = np.argmax(probabilities, axis=1)

        y_true_perturbed.extend(labels.numpy())
        y_pred_perturbed.extend(predictions)

    robustness_accuracy = accuracy_score(
        y_true_perturbed,
        y_pred_perturbed
    )

    robustness_precision = precision_score(
        y_true_perturbed,
        y_pred_perturbed,
        average="macro",
        zero_division=0
    )

    robustness_recall = recall_score(
        y_true_perturbed,
        y_pred_perturbed,
        average="macro",
        zero_division=0
    )

    robustness_f1 = f1_score(
        y_true_perturbed,
        y_pred_perturbed,
        average="macro",
        zero_division=0
    )

    return {
        "accuracy": robustness_accuracy,
        "precision_macro": robustness_precision,
        "recall_macro": robustness_recall,
        "f1_macro": robustness_f1
    }


noise_robustness = evaluate_on_perturbed_dataset(
    model=model,
    dataset=test_ds,
    perturbation_type="noise"
)

brightness_robustness = evaluate_on_perturbed_dataset(
    model=model,
    dataset=test_ds,
    perturbation_type="brightness"
)

noise_accuracy_drop = test_accuracy - noise_robustness["accuracy"]
brightness_accuracy_drop = test_accuracy - brightness_robustness["accuracy"]

print("\n================ ROBUSTNESS TESTING ================")

print("\nGaussian Noise Robustness:")
print(f"Accuracy        : {noise_robustness['accuracy']:.4f}")
print(f"Precision Macro : {noise_robustness['precision_macro']:.4f}")
print(f"Recall Macro    : {noise_robustness['recall_macro']:.4f}")
print(f"F1 Macro        : {noise_robustness['f1_macro']:.4f}")
print(f"Accuracy Drop   : {noise_accuracy_drop:.4f}")

print("\nBrightness Perturbation Robustness:")
print(f"Accuracy        : {brightness_robustness['accuracy']:.4f}")
print(f"Precision Macro : {brightness_robustness['precision_macro']:.4f}")
print(f"Recall Macro    : {brightness_robustness['recall_macro']:.4f}")
print(f"F1 Macro        : {brightness_robustness['f1_macro']:.4f}")
print(f"Accuracy Drop   : {brightness_accuracy_drop:.4f}")


# ============================================================
# 25. GRAD-CAM EXPLAINABILITY ANALYSIS
# ============================================================

def get_last_conv_layer_name(model):
    """
    Automatically identifies the last Conv2D layer in the model.
    """

    for layer in reversed(model.layers):
        if isinstance(layer, tf.keras.layers.Conv2D):
            return layer.name

    raise ValueError("No Conv2D layer found in the model.")


LAST_CONV_LAYER_NAME = get_last_conv_layer_name(model)

print("\n================ GRAD-CAM CONFIGURATION ================")
print("Last convolutional layer:", LAST_CONV_LAYER_NAME)


def make_gradcam_heatmap(
    img_array,
    model,
    last_conv_layer_name,
    pred_index=None
):
    """
    Generates a Grad-CAM heatmap for a given input image.
    """

    grad_model = tf.keras.models.Model(
        inputs=model.inputs,
        outputs=[
            model.get_layer(last_conv_layer_name).output,
            model.output
        ]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)

        if pred_index is None:
            pred_index = tf.argmax(predictions[0])

        class_channel = predictions[:, pred_index]

    gradients = tape.gradient(class_channel, conv_outputs)

    pooled_gradients = tf.reduce_mean(
        gradients,
        axis=(0, 1, 2)
    )

    conv_outputs = conv_outputs[0]

    heatmap = conv_outputs @ pooled_gradients[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    max_heatmap_value = tf.reduce_max(heatmap)

    if max_heatmap_value == 0:
        heatmap = tf.zeros_like(heatmap)
    else:
        heatmap = tf.maximum(heatmap, 0) / max_heatmap_value

    return heatmap.numpy()


def superimpose_gradcam(img, heatmap, alpha=0.40):
    """
    Superimposes the Grad-CAM heatmap on the original image.
    """

    img = np.array(img).astype(np.uint8)

    heatmap = cv2.resize(
        heatmap,
        (img.shape[1], img.shape[0])
    )

    heatmap = np.uint8(255 * heatmap)

    jet = cm.get_cmap("jet")
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]

    jet_heatmap = np.uint8(255 * jet_heatmap)

    superimposed_img = cv2.addWeighted(
        img,
        1 - alpha,
        jet_heatmap,
        alpha,
        0
    )

    return superimposed_img


def generate_gradcam_samples(
    model,
    dataset,
    class_names,
    last_conv_layer_name,
    output_dir,
    num_images=10
):
    """
    Generates and saves Grad-CAM visualizations for sample test images.
    """

    saved_count = 0

    for images, labels in dataset:

        probabilities = model.predict(images, verbose=0)
        predictions = np.argmax(probabilities, axis=1)

        for i in range(images.shape[0]):

            if saved_count >= num_images:
                print(f"\nGenerated {saved_count} Grad-CAM samples.")
                return saved_count

            img = images[i].numpy().astype(np.uint8)
            img_array = np.expand_dims(images[i].numpy(), axis=0)

            true_label_index = int(labels[i].numpy())
            predicted_label_index = int(predictions[i])

            true_label = class_names[true_label_index]
            predicted_label = class_names[predicted_label_index]
            confidence = float(np.max(probabilities[i]))

            heatmap = make_gradcam_heatmap(
                img_array=img_array,
                model=model,
                last_conv_layer_name=last_conv_layer_name,
                pred_index=predicted_label_index
            )

            overlay_img = superimpose_gradcam(
                img=img,
                heatmap=heatmap,
                alpha=0.40
            )

            safe_true_label = true_label.replace(" ", "_").replace("/", "_")
            safe_predicted_label = predicted_label.replace(" ", "_").replace("/", "_")

            sample_filename = (
                f"sample_{saved_count + 1}_"
                f"true_{safe_true_label}_"
                f"pred_{safe_predicted_label}_"
                f"conf_{confidence:.2f}"
            )

            original_path = os.path.join(
                output_dir,
                sample_filename + "_original.png"
            )

            heatmap_path = os.path.join(
                output_dir,
                sample_filename + "_heatmap.png"
            )

            overlay_path = os.path.join(
                output_dir,
                sample_filename + "_overlay.png"
            )

            plt.imsave(original_path, img)
            plt.imsave(heatmap_path, heatmap, cmap="jet")
            plt.imsave(overlay_path, overlay_img)

            saved_count += 1

    print(f"\nGenerated {saved_count} Grad-CAM samples.")
    return saved_count


def display_gradcam_results(
    model,
    dataset,
    class_names,
    last_conv_layer_name,
    num_images=5
):
    """
    Displays original image, Grad-CAM heatmap, and overlay image
    for selected test images in Google Colab.
    """

    displayed_count = 0

    plt.figure(figsize=(15, num_images * 4))

    for images, labels in dataset:

        probabilities = model.predict(images, verbose=0)
        predictions = np.argmax(probabilities, axis=1)

        for i in range(images.shape[0]):

            if displayed_count >= num_images:
                plt.tight_layout()
                plt.show()
                return

            img = images[i].numpy().astype(np.uint8)
            img_array = np.expand_dims(images[i].numpy(), axis=0)

            true_label_index = int(labels[i].numpy())
            predicted_label_index = int(predictions[i])

            true_label = class_names[true_label_index]
            predicted_label = class_names[predicted_label_index]
            confidence = float(np.max(probabilities[i]))

            heatmap = make_gradcam_heatmap(
                img_array=img_array,
                model=model,
                last_conv_layer_name=last_conv_layer_name,
                pred_index=predicted_label_index
            )

            overlay_img = superimpose_gradcam(
                img=img,
                heatmap=heatmap,
                alpha=0.40
            )

            row = displayed_count * 3

            plt.subplot(num_images, 3, row + 1)
            plt.imshow(img)
            plt.title(f"Original\nTrue: {true_label}")
            plt.axis("off")

            plt.subplot(num_images, 3, row + 2)
            plt.imshow(heatmap, cmap="jet")
            plt.title("Grad-CAM Heatmap")
            plt.axis("off")

            plt.subplot(num_images, 3, row + 3)
            plt.imshow(overlay_img)
            plt.title(
                f"Overlay\nPred: {predicted_label}\nConfidence: {confidence:.2f}"
            )
            plt.axis("off")

            displayed_count += 1

    plt.tight_layout()
    plt.show()


NUM_GRADCAM_SAMPLES = 10

gradcam_samples_generated = generate_gradcam_samples(
    model=model,
    dataset=test_ds,
    class_names=class_names,
    last_conv_layer_name=LAST_CONV_LAYER_NAME,
    output_dir=GRADCAM_OUTPUT_DIR,
    num_images=NUM_GRADCAM_SAMPLES
)

display_gradcam_results(
    model=model,
    dataset=test_ds,
    class_names=class_names,
    last_conv_layer_name=LAST_CONV_LAYER_NAME,
    num_images=5
)


# ============================================================
# 26. SAVE COMPLETE EVALUATION RESULTS TO JSON
# ============================================================

evaluation_results = {
    "model_information": {
        "experiment_group": EXPERIMENT_GROUP,
        "model_name": MODEL_NAME,
        "model_type": "Custom CNN with DropBlock2D",
        "comparison_reference": "G1 Custom CNN without Dropout",
        "dropout_type": "DropBlock2D",
        "dropout_rate": float(DROPBLOCK_RATE),
        "dropblock_block_size": int(DROPBLOCK_BLOCK_SIZE),
        "dropout_placement": "After deeper convolutional blocks: Block 3 and Block 4",
        "image_size": f"{IMG_HEIGHT}x{IMG_WIDTH}",
        "batch_size": BATCH_SIZE,
        "epochs_configured": EPOCHS,
        "epochs_completed": len(history.history["loss"]),
        "number_of_classes": NUM_CLASSES,
        "class_names": class_names
    },

    "test_metrics": {
        "accuracy": float(test_accuracy),
        "precision_macro": float(test_precision_macro),
        "recall_macro": float(test_recall_macro),
        "f1_macro": float(test_f1_macro),
        "precision_weighted": float(test_precision_weighted),
        "recall_weighted": float(test_recall_weighted),
        "f1_weighted": float(test_f1_weighted)
    },

    "generalization_gap": {
        "best_training_accuracy": float(best_train_accuracy),
        "best_validation_accuracy": float(best_val_accuracy),
        "best_generalization_gap": float(best_generalization_gap),
        "final_training_accuracy": float(final_train_accuracy),
        "final_validation_accuracy": float(final_val_accuracy),
        "final_generalization_gap": float(final_generalization_gap)
    },

    "robustness_testing": {
        "gaussian_noise": {
            "accuracy": float(noise_robustness["accuracy"]),
            "precision_macro": float(noise_robustness["precision_macro"]),
            "recall_macro": float(noise_robustness["recall_macro"]),
            "f1_macro": float(noise_robustness["f1_macro"]),
            "accuracy_drop": float(noise_accuracy_drop)
        },

        "brightness_perturbation": {
            "accuracy": float(brightness_robustness["accuracy"]),
            "precision_macro": float(brightness_robustness["precision_macro"]),
            "recall_macro": float(brightness_robustness["recall_macro"]),
            "f1_macro": float(brightness_robustness["f1_macro"]),
            "accuracy_drop": float(brightness_accuracy_drop)
        }
    },

    "model_complexity": {
        "total_parameters": int(total_params),
        "model_size_mb": float(model_size_mb)
    },

    "inference_speed": {
        "average_inference_time_per_image_seconds": float(avg_inference_time),
        "images_per_second": float(images_per_second)
    },

    "training_time": {
        "training_time_seconds": float(training_time_seconds)
    },

    "explainability": {
        "method": "Grad-CAM",
        "last_convolutional_layer": LAST_CONV_LAYER_NAME,
        "number_of_gradcam_samples_generated": int(gradcam_samples_generated),
        "gradcam_output_directory": GRADCAM_OUTPUT_DIR,
        "purpose": (
            "Grad-CAM was used to visualize the image regions that contributed "
            "most strongly to the CNN classification decision. The visual explanation "
            "supports the assessment of whether the model focused on disease-relevant "
            "regions such as spots, lesions, discoloration, necrotic patches, and "
            "texture irregularities."
        )
    }
}

with open(RESULTS_JSON_PATH, "w") as json_file:
    json.dump(evaluation_results, json_file, indent=4)

print("\n================ JSON RESULTS SAVED ================")
print(f"Evaluation results saved to: {RESULTS_JSON_PATH}")


# ============================================================
# 27. SUMMARY TABLE FOR DISSERTATION REPORTING
# ============================================================

summary_table = pd.DataFrame({
    "Metric": [
        "Experiment Group",
        "Model Name",
        "Comparison Reference",
        "Dropout Type",
        "Dropout Rate",
        "DropBlock Block Size",
        "Dropout Placement",
        "Accuracy",
        "Precision Macro",
        "Recall Macro",
        "F1-score Macro",
        "Precision Weighted",
        "Recall Weighted",
        "F1-score Weighted",
        "Best Generalization Gap",
        "Final Generalization Gap",
        "Noise Accuracy Drop",
        "Brightness Accuracy Drop",
        "Total Parameters",
        "Model Size MB",
        "Average Inference Time per Image",
        "Images per Second",
        "Training Time Seconds",
        "Explainability Method",
        "Grad-CAM Layer",
        "Grad-CAM Samples Generated"
    ],

    "Value": [
        EXPERIMENT_GROUP,
        MODEL_NAME,
        "G1 Custom CNN without Dropout",
        "DropBlock2D",
        DROPBLOCK_RATE,
        DROPBLOCK_BLOCK_SIZE,
        "After Conv Blocks 3 and 4",
        test_accuracy,
        test_precision_macro,
        test_recall_macro,
        test_f1_macro,
        test_precision_weighted,
        test_recall_weighted,
        test_f1_weighted,
        best_generalization_gap,
        final_generalization_gap,
        noise_accuracy_drop,
        brightness_accuracy_drop,
        total_params,
        model_size_mb,
        avg_inference_time,
        images_per_second,
        training_time_seconds,
        "Grad-CAM",
        LAST_CONV_LAYER_NAME,
        gradcam_samples_generated
    ]
})

summary_table.to_csv(SUMMARY_TABLE_PATH, index=False)

print("\n================ SUMMARY TABLE ================")
print(summary_table)

print(f"\nSummary table saved to: {SUMMARY_TABLE_PATH}")


# ============================================================
# 28. FINAL OUTPUT SUMMARY
# ============================================================

print("\n============================================================")
print("G4 CUSTOM CNN WITH DROPBLOCK COMPLETED SUCCESSFULLY")
print("============================================================")

print("\nSaved Outputs:")
print(f"1. Final model             : {MODEL_SAVE_PATH}")
print(f"2. Best model              : {BEST_MODEL_SAVE_PATH}")
print(f"3. Training log            : {TRAINING_LOG_PATH}")
print(f"4. Classification report   : {CLASSIFICATION_REPORT_PATH}")
print(f"5. Confusion matrix CSV    : {CONFUSION_MATRIX_PATH}")
print(f"6. Evaluation results JSON : {RESULTS_JSON_PATH}")
print(f"7. Summary table           : {SUMMARY_TABLE_PATH}")
print(f"8. Grad-CAM outputs folder : {GRADCAM_OUTPUT_DIR}")

print("\nKey Results:")
print(f"Experiment Group               : {EXPERIMENT_GROUP}")
print(f"Model Name                     : {MODEL_NAME}")
print(f"Dropout Type                   : DropBlock2D")
print(f"Dropout Rate                   : {DROPBLOCK_RATE}")
print(f"DropBlock Block Size           : {DROPBLOCK_BLOCK_SIZE}")
print(f"Dropout Placement              : After Conv Blocks 3 and 4")
print(f"Accuracy                       : {test_accuracy:.4f}")
print(f"Precision Macro                : {test_precision_macro:.4f}")
print(f"Recall Macro                   : {test_recall_macro:.4f}")
print(f"F1-score Macro                 : {test_f1_macro:.4f}")
print(f"Best Generalization Gap        : {best_generalization_gap:.4f}")
print(f"Noise Accuracy Drop            : {noise_accuracy_drop:.4f}")
print(f"Brightness Accuracy Drop       : {brightness_accuracy_drop:.4f}")
print(f"Total Parameters               : {total_params:,}")
print(f"Model Size                     : {model_size_mb:.2f} MB")
print(f"Average Inference Time/Image   : {avg_inference_time:.6f} seconds")
print(f"Images per Second              : {images_per_second:.2f}")
print(f"Grad-CAM Layer                 : {LAST_CONV_LAYER_NAME}")
print(f"Grad-CAM Samples Generated     : {gradcam_samples_generated}")

```
